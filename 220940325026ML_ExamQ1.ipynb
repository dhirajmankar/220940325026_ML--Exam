```python
df1=df.drop('Car_Name',axis=1)
```

```python
#This cell can only be run once
Fuel_Type=df1[['Fuel_Type']]
Fuel_Type =pd.get_dummies(Fuel_Type, drop_first=True)

Seller_Type=df1[['Seller_Type']]
Seller_Type=pd.get_dummies(Seller_Type, drop_first=True)


Transmission=df1[['Transmission']]
Transmission=pd.get_dummies(Transmission, drop_first=True)

df1=pd.concat([df1,Fuel_Type, Seller_Type, Transmission], axis=1)

df1.drop(labels=['Fuel_Type', 'Seller_Type', 'Transmission'], axis=1, inplace=True)

df1.head()
```

```python
sell =df1['Selling_Price']
df1.drop(['Selling_Price'],axis=1,inplace = True)
df1=df1.join(sell)
```

```python
x= df1.iloc[:,:-1].values
y = df1.iloc[:,-1].values
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(random_state = 0)
dt_reg.fit(x_train, y_train)
y_pred=dt_reg.predict(x_test)
```
