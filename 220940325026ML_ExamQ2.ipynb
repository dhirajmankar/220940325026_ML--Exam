```python
#independant values
x = df.iloc[:,1:-1].values
# dependant variable 
y = df.iloc[:,-1].values
x=x.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#As it is given we will use Polynomial Regression
# importing liraries
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
regrr=LinearRegression()
regrr.fit(x,y)
```

```python
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
```

```python
regrr2=LinearRegression()
regrr2.fit(x_poly,y)
```

```python
#Visualizing the results of our linear model1
plt.scatter(x,y,color='yellow')
plt.plot(x,regrr.predict(x),color='red')
plt.title('LinearRegression1')
plt.xlabel('Features')
plt.ylabel('Price')
```

```python
#Visualizing the result of Polynomial regression
plt.scatter(x,y,color='green')
plt.plot(x,regrr2.predict(poly.fit_transform(x)),color='red')
plt.title('PolynomialRegression1')
plt.xlabel('Features')
plt.ylabel('Price')
```

```python
poly2=PolynomialFeatures(degree=4)
x_poly2=poly2.fit_transform(x)
```

```python
regr3=LinearRegression()
regr3.fit(x_poly2,y)
```

```python
#Visualizing the results of our linear model2
plt.scatter(x,y,color='yellow')
plt.plot(x,regr3.predict(x),color='red')
plt.title('LinearRegression2')
plt.xlabel('Features')
plt.ylabel('Price')
```

```python
#Visualizing the result of Polynomial regression2
plt.scatter(x,y,color='green')
plt.plot(x,regr3.predict(poly2.fit_transform(x)),color='red')
plt.title('PolynomialRegression2')
plt.xlabel('Features')
plt.ylabel('Price')
```
